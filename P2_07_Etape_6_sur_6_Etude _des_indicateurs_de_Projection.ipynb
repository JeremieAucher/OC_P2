<p><u>Sélection de l'Indicator Code</u>:</p>

<p>Après lecture et analyse des 102 indicateurs, je sélectionne l'indicateur <b>PRJ.ATT.ALL.4.MF</b></p>

<p><u>Définition</u>: 	Share of the population of the stated age group that has completed post-secondary or tertiary education as the highest level of educational attainment. Projections are based on collected census and survey data for the base year (around 2010) and the Medium Shared Socioeconomic Pathways (SSP2) projection model. The SSP2 is a middle-of-the-road scenario that combines medium fertility with medium mortality, medium migration, and the Global Education Trend (GET) education scenario.</p>

<p>Taux de remplissage de l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>'</p>

```python
data[['Indicator Code','Taux de Remplissage']][data['Indicator Code'] == 'PRJ.ATT.ALL.4.MF'].groupby('Indicator Code').mean()
```

<p><u>Import de la liste des pays selectionnés</u> sur la base des 3 indicateurs: '<b>IT.NET.USER.P2</b>','<b>NY.GDP.PCAP.CD</b>' et '<b>popTotalEtudiant</b>'</p>

```python
listePaysFinalistes = np.array(['United States', 'Japan', 'United Kingdom', 'Germany',
       'Korea, Rep.', 'France', 'Spain', 'Canada', 'Australia',
       'Netherlands', 'Belgium', 'Sweden', 'Austria'], dtype=object)
```

<p><u>Enregistrement de la moyenne de référence</u> :</p>
<p>Moyenne calculée à partir des pays finalistes <br \>
    sur l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>' sur <br \>
    <u>l'année <b>2015</b></u> (dernière date disponible qui <br \>ne soit pas une projection).</p>

```python
valeurMoyRef = data['2015'][(data['Indicator Code'] == 'PRJ.ATT.ALL.4.MF') \
                            & (data['Country Name'].isin(listePaysFinalistes))].mean()
print('Valeur arrondi de la Moyenne de Reference :{}'.format(round(valeurMoyRef,2)))
```

<h4>Data Cleaning du DF 'data':</h4>
<p>
    <ol>
        <li>Suppression des pays sans valeur renseignée</li>
        <li>Suppression des '<i>Indicator Code</i>' à l'exception de '<b>PRJ.ATT.ALL.4.MF</b>'</li>
        <li>Suppression des colonnes inutiles</li>
    </ol>
</p>

<h5>Suppression des pays sans valeur renseignée</h5>

<p>Affichage des Pays avec leur moyenne de 'Taux de Remplissage'</p>

```python
data[['Country Name','Taux de Remplissage']].groupby('Country Name').mean()
```

<p>Suppression des pays qui ont une moyenne de '<b>Taux de Remplissage</b>' égale à 0</p>

```python
dataPaysVide = data[['Country Name','Taux de Remplissage']].groupby('Country Name').mean()
dataPaysVide = dataPaysVide[dataPaysVide['Taux de Remplissage'] == 0].index
print(dataPaysVide)
data = data[~data['Country Name'].isin(dataPaysVide)]
```

<h5>Suppression des 'Indicator Code' à l'exception de 'PRJ.ATT.ALL.4.MF'</h5>

```python
data = data[data['Indicator Code'] == 'PRJ.ATT.ALL.4.MF']
data.head()
```

<h5>Suppression des colonnes inutiles</h5>

```python
data.drop(columns=['Indicator Code','Topic','Taux de Remplissage'], inplace=True)
data.head()
```

<p>Il n'est pas très utile pour notre projet de regarder trop loin dans le futur.<br \>
    Je décide de <u>ne pas regarder au-delà de l'horizon <b>2030</b></u> pour détecter <br \>
    des pays qui pourront avoir du potentiel dans un futur assez proche.<br \>
    <u>Je supprime donc également les colonnes représentant les années <br \>
        au-delà de l'année 2030</u>.</p>

```python
data.drop(columns=data.loc[:,'2035':].columns, inplace=True)
data.head()
```

<h4>Feature Modeling du DF 'data':</h4>

<p><u>Ajout de 2 colonnes</u> '<b>valeurMin</b>' et '<b>valeurMax</b>' qui indiqueront pour chaque pays, <br \>
    leur <u>valeur minimale et maximale</u>, sur la période <b>2010 à 2030</b>, pour l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>'.</p>

```python
data['valeurMin'] = data.loc[:,'2010':].apply(lambda row: min(row), axis=1)
data['valeurMax'] = data.loc[:,'2010':].apply(lambda row: max(row), axis=1)
data.head()
```

<h4>Data Visualisation</h4>

<p>Mise en graphique de l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>':
    <ol>
        <li>Graphique en courbe (plot)</li>
        <li>Représente, pour chaque pays, le taux de sa population ayant<br \> 
            terminé des études postsecondaires ou supérieures comme <br \>
            niveau d'instruction le plus élevé</li>
        <li>Les valeurs sont affichées <u>tous les 5 ans</u> de <b>2010</b> à <b>2030</b></li>

        <li>Les pays sont regroupés par région (1 graphique / région)</li>
    </ol>
</p>
<p>L'objectif est d'observer les pays qui aujourd'hui ont un taux, <br \>
    pour l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>', inférieur à la moyenne <br \>
    de référence mais qui auront un taux supérieur à celle-ci d'ici l'année <b>2030</b>.<br \>
Ces pays seront ensuite sélectionnés avant une dernière étape <br \>

<p>Affiche par <b>région</b> puis par <b>pays</b> les valeurs de l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>'</p>

```python
fig = plt.figure(figsize=(15,30))
x = np.array(data.loc[:,'2010':"2030"].columns)

for idx, region in enumerate(data['Region'][~data['Country Name'].isin(listePaysFinalistes)].unique(), start=1):
    ax = fig.add_subplot(4,2,idx)
    ax.set_xlabel('PRJ.ATT.ALL.4.MF')

    plt.xticks(rotation=45)
    plt.ylim(0,1)
    plt.title(region)
    plt.plot(x, np.linspace(valeurMoyRef,valeurMoyRef,len(x)), 'k-', lw=2)

    for country in data['Country Name'][(data['Region'] == region)&(~data['Country Name'].isin(listePaysFinalistes))].unique():    
        plt.plot(x,data[x][data['Country Name'] == country].iloc[0])
plt.tight_layout(pad=3.0)
```

<p>Nombre de pays dont la valeur de l'indicateur '<b>PRJ.ATT.ALL.4.MF</b>' <br \>
    dépasse la <i>moyenne de référence</i> sur <u>la période 2010 à 2030</u> :</p>

```python
len(data[(data['valeurMax']>valeurMoyRef)&(~data['Country Name'].isin(listePaysFinalistes))]['Country Name'].unique())
```

<p>Detail des pays regroupés par région : </p>

```python
data[(data['valeurMax']>valeurMoyRef)&(~data['Country Name'].isin(listePaysFinalistes))][['Country Name','Region']].groupby('Region').count().sort_values(by=['Country Name'], ascending=False)
```

<p>Détail des 15 pays séléctionnés :<br \>

```python
data[(data['valeurMax']>valeurMoyRef)&(~data['Country Name']\
    .isin(listePaysFinalistes))][['Country Name','Region']]\
    .sort_values(by=['Region','Country Name'], ascending=[True, True])
```

<p>Je vais maintenant finaliser cette étude en testant ces <u><b>15</b> pays</u> <br \>

<p><u>Les indicateurs utilisés seront</u> :

<p>L'objectif est de réaliser, comme à l'étape précédente, <br \>

    dont les caractéristiques relatives aux indicateurs sélectionnés, <br \>

<p><u>Détail des dernières étapes</u> :
    <ol>
        <li>Sauvegarde dans un <i>array</i> '<b>listePaysPRJ</b>' la liste des pays sélectionnés</li>
        <li>Importation du DF '<b>dataPivotBackup</b>' via <i>Pickle</i> dans le DF '<b>dataITGDPPop</b>'</li>
        <li>Création du pays de référence des pays finalistes : '<b>paysFinalistes</b>'
                    <ol>
                        <li>Récupération des valeurs moyennes des pays finalistes pour les indicateurs '<b>IT.NET.USER.P2</b>' et '<b>NY.GDP.PCAP.CD</b>'</li>

        <li>Filtrage du DF '<b>dataITGDPPop</b>' avec la liste '<b>listePays</b>'</li>

        <li><u>Data Visualisation</u> :
            <ol><li>Mise en graphique des pays sélectionnés et du pays représentant les pays finalistes '<b>paysFinalistes</b>'</li></ol></li>

    </ol>
</p>

<p>Sauvegarde dans un <i>array</i> '<b>listePaysPRJ</b>' la liste des pays sélectionnés</p>

```python
listePaysPRJ = data[(data['valeurMax']>valeurMoyRef)&
                    (~data['Country Name'].isin(listePaysFinalistes))]\
                    ['Country Name'].unique()
listePaysPRJ
```

<p>Importation du DF '<b>dataPivotBackup</b>' via <i>Pickle</i> dans le DF '<b>dataITGDPPop</b>'</p>

```python
dataITGDPPop = pickle.load(open('dataPivotBackup', 'rb'))
```

<p><u>Création du pays de référence des pays finalistes: '<b>paysFinalistes</b>'</u></p>
<p>
    <ol>
        <li>Récupération des valeurs moyennes des pays pays finalistes pour les indicateurs '<b>IT.NET.USER.P2</b>' et '<b>NY.GDP.PCAP.CD</b>'</li>
        <li>Récupération de la valeur minimale des pays finalistes pour l'indicateur '<b>popTotalEtudiant</b>'</li>
        <li>Création du nouveau pays '<b>paysFinalistes</b>' via la création d'un DF du même nom ayant la même structure que le DF '<b>dataITGDPPop</b>'</li>
    </ol>
</p>

<p>Récupération des valeurs moyennes des pays finalistes pour les indicateurs '<b>IT.NET.USER.P2</b>' et '<b>NY.GDP.PCAP.CD</b>'</p>

```python
print('\nValeur moyenne pour \'IT.NET.USER.P2\': {}'\
      .format(round(dataITGDPPop['IT.NET.USER.P2']\
            [dataITGDPPop['Country Name'].isin(listePaysFinalistes)].mean(),2)))
print('Valeur moyenne pour \'NY.GDP.PCAP.CD\': {}'\
      .format(round(dataITGDPPop['NY.GDP.PCAP.CD']\
            [dataITGDPPop['Country Name'].isin(listePaysFinalistes)].mean(),2)))

moyIT = round(dataITGDPPop['IT.NET.USER.P2']\
            [dataITGDPPop['Country Name'].isin(listePaysFinalistes)].mean(),2)
moyGDP = round(dataITGDPPop['NY.GDP.PCAP.CD']\
            [dataITGDPPop['Country Name'].isin(listePaysFinalistes)].mean(),2)
```

<p>Récupération de la valeur minimale des pays finalistes pour l'indicateur '<b>popTotalEtudiant</b>'</p>

```python
print('Valeur minimale pour \'popTotalEtudiant\': {}'\
      .format(int(dataITGDPPop['popTotalEtudiant']\
            [dataITGDPPop['Country Name'].isin(listePaysFinalistes)].min())))

minPop = int(dataITGDPPop[dataITGDPPop['Country Name'].isin(listePaysFinalistes)]['popTotalEtudiant'].min())
```

<p>Création du nouveau pays '<b>paysFinalistes</b>' via la création d'un DF <br \>
    du même nom ayant la même structure que le DF '<b>data</b>'</p>

```python
paysFinalistes = pd.DataFrame([['paysFinalistes',
               'NA',
               moyIT,
               moyGDP,
               minPop]],
             columns=['Country Name',
                      'Region',
                      'IT.NET.USER.P2',
                      'NY.GDP.PCAP.CD',
                      'popTotalEtudiant'])
paysFinalistes
```

<p>Filtrage du DF '<b>dataITGDPPop</b>' avec la liste '<b>listePaysPRJ</b>'</p>

```python
dataITGDPPop = dataITGDPPop[dataITGDPPop['Country Name'].isin(listePaysPRJ)]
```

<p>Ajout du pays '<b>paysFinalistes</b>' comme nouvelle entrée au DF '<b>dataITGDPPop</b>'</p>

```python
dataITGDPPop = dataITGDPPop.append(paysFinalistes)
```

<h4>Data Visualisation</h4>

<p>Mise en graphique des pays sélectionnés et du pays représentant les pays finalistes '<b>paysFinalistes</b>'</p>

```python
rl = sns.relplot(
        x="NY.GDP.PCAP.CD", 
        y="IT.NET.USER.P2", 
        hue="Country Name", 
        size="popTotalEtudiant",
        sizes=(40,400),
        alpha=.5, 
        palette="muted",
        height=6, 
        data=dataITGDPPop)
axes = rl.axes
# axes[0,0].set_xlim(0,minMaxIndicFix[2][1])
# axes[0,0].set_ylim(minMaxIndicFix[0][0],100)
axes[0,0].set_title('Monde')
```

<p>Je sélectionne les pays ayant :
    <ul>

        <li>une population supérieure à 500 000 étudiants <i>(secondaire sup + tertiaire)</i></li>

```python
# Liste finalisée des pays à futur potentiel intéressants
dataITGDPPop[(dataITGDPPop['popTotalEtudiant'] > 500000) & 
             (dataITGDPPop['NY.GDP.PCAP.CD'] > 35000)]
```

```python
rl = sns.relplot(
        x="NY.GDP.PCAP.CD", 
        y="IT.NET.USER.P2", 
        hue="Country Name", 
        size="popTotalEtudiant",
        sizes=(40,400),
        alpha=.5, 
        palette="muted",
        height=6, 
        data=dataITGDPPop[(dataITGDPPop['popTotalEtudiant'] > 500000) & 
             (dataITGDPPop['NY.GDP.PCAP.CD'] > 35000)])
axes = rl.axes
axes[0,0].set_xlim(30000,72000)
axes[0,0].set_ylim(85,100)
axes[0,0].set_title('Monde')
```
